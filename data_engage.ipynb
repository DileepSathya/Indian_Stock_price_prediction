```python
import pandas as pd
import os
colums=[5,6,9,10,11,12,13,14]
final=pd.DataFrame()
path = "C://Users//Dileep Sathya//OneDrive//Desktop//projects//stock price prediction//new_EQ_data"
dir_list = os.listdir(path)


for file in dir_list:
    df=pd.read_csv('new_EQ_data//'+file)


    # Specify the row and column to shift
    for cols in colums:
            column_index = cols  # Column B corresponds to index 1 (0-based index)
            start_row = 0  # Specify the row (5th row in 0-based index)
            shift_by = 1  # Specify the number of rows to shift down

            # Shift the specified cell and all cells below it down by the specified amount
            df.iloc[start_row + shift_by:, column_index] = df.iloc[start_row:-shift_by, column_index].values

            # Clear the original cell
            df.iloc[start_row, column_index] = None
    df=df.dropna()
    final=pd.concat([df,final],axis=0)
```
