make the data set so that only eq series will be there

```python
import pandas as pd
import os
path = "C://Users//Dileep Sathya//OneDrive//Desktop//projects//stock price prediction//data"
dir_list = os.listdir(path)
for file in dir_list:

    df=pd.read_csv('data//'+file)
    #removing extra spaces in column names
    df.columns = df.columns.str.strip()
    df=df[df['Series'] =='EQ']
    df.to_csv('new_EQ_data//'+file,index=False)
```

 feature engineering

```python
path2="C://Users//Dileep Sathya//OneDrive//Desktop//projects//stock price prediction//new_EQ_data"
dir_list_2=os.listdir(path2)
for file in dir_list_2:

    df2=pd.read_csv('new_EQ_data//'+file)
    for col in df2.columns:
        if col not in ['Symbol', 'Date', 'Series', 'prev_Dly_Qt_to_Traded_Qty']:
            if df2[col].dtype == 'object':
                df2[col] = df2[col].apply(lambda x: str(x).replace(',', '') if isinstance(x, str) else x)

                df2[col] = pd.to_numeric(df2[col], errors='coerce')
```
